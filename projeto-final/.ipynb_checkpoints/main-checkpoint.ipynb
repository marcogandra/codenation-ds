### Função para saber mais sobre o dataframe

```python
def exploracao(df):
    exploracao = pd.DataFrame({
        'coluna' : df.columns,
        'tipos' : df.dtypes,
        'qtd_null' : df.isnull().sum()
    })
    exploracao['percent_null'] = np.round((exploracao.qtd_null / df.shape[0]) * 100, 2)
    
    return exploracao.sort_values(by='qtd_null', ascending=False)
```
